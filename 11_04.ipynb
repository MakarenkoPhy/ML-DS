```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
cancer = load_breast_cancer()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
print(mlp.score(X_test, y_test))
```

```python
y_probas = mlp.predict_proba(X_test)
skplt.metrics.plot_precision_recall(y_test, y_probas)
plt.show()
```

```python
#print(y_probas[:, 0])
yy=y_probas[:, 1]
print(yy>0.7)
```

```python
cm = confusion_matrix(y_test, yy>0.5)
f = sns.heatmap(cm, annot=True)
plt.ylabel('Groud truth')
plt.xlabel('Prediction')
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import auc
prs, recs, _ = precision_recall_curve(y_test, y_probas[:, 1])
plt.plot(recs, prs, marker='o')
plt.ylim([0, 1.1]);plt.xlim([0, 1.1])
plt.xlabel('Recall');plt.ylabel('Precision')
plt.title('PR_curve')
auc(recs, prs)
```

```python
from sklearn.metrics import plot_roc_curve, roc_curve
plot_roc_curve(mlp, X_train, y_train)
plt.ylim([0, 1.1]);
```
